```python
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample_rate, data = wavfile.read("waveform_exercise.wav")
sample_rate, data
```

```python
data = data[:, 0] # Left channel 
```

```python
n = len(data)
yf = rfft(data)
xf = rfftfreq(n, 1 / sample_rate)

yf, xf, n
```

```python
# 4. Get magnitude and filter for positive frequencies
# Only the first half of the FFT results are needed for real signals
magnitude = np.abs(yf)
frequencies = xf
```

```python
plt.figure(figsize=(16, 4))
plt.plot(frequencies, magnitude)
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
